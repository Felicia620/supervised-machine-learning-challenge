```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
train_df_numeric = pd.get_dummies(train_df)
train_df_numeric
```

```python
y_train = train_df_numeric["loan_status_low_risk"]
y_train
```

```python
train_df_numeric.drop(["loan_status_low_risk"], axis = 1, inplace = True)
train_df_numeric.drop(["loan_status_high_risk"], axis = 1, inplace = True)
```

```python
X_train = train_df_numeric
X_train.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_df_numeric = pd.get_dummies(test_df)
test_df_numeric
```

```python
y_test = test_df_numeric["loan_status_low_risk"]
y_test
```

```python
test_df_numeric.drop(["loan_status_low_risk"], axis = 1, inplace = True)
test_df_numeric.drop(["loan_status_high_risk"], axis = 1, inplace = True)
```

```python
X_test = test_df_numeric
X_test.head()
```

```python
# add missing dummy variables to testing set
for columns in X_train.columns:
    if columns not in X_test.columns:
        X_test[columns] = 0
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver = "sag", max_iter = 100)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)

print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
clf_scaled = LogisticRegression().fit(X_train_scaled, y_train)
print(f"Training Score: {clf_scaled.score(X_train_scaled, y_train)}")
print(f"Testing Score: {clf_scaled.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
rf_clf_scaled = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f"Training Score: {rf_clf_scaled.score(X_train_scaled, y_train)}")
print(f"Testing Score: {rf_clf_scaled.score(X_test_scaled, y_test)}")
```
